```python
#importing necessary libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading the csv file
data = pd.read_csv("Car data.csv")
data
```

```python
# Getting info of the dataset

data.info()
```

```python
# Checking if there is any null value or not

data.isnull().sum()
```

```python
# Getting all the unique car names

data.Car_Name.unique()
```

```python
# Count the occurrences of each car name
car_name_counts = data['Car_Name'].value_counts()

# Plot a pie chart
plt.figure(figsize=(10, 10))
plt.pie(car_name_counts, labels=car_name_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Car Names')
plt.axis('equal')  
plt.show()
```

```python
# Count the occurrences of each fuel type
fuel_type_counts = data['Fuel_Type'].value_counts()

# Plot a pie chart
plt.figure(figsize=(10, 10))
plt.pie(fuel_type_counts, labels=fuel_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of fuel types')
plt.axis('equal')  
plt.show()
```

```python
# Counting the occurrences of each unique combination of car name and year

car_year_counts = data.groupby(['Car_Name', 'Year']).size().reset_index(name='Count')

# Plot the bar chart for Car_Name vs. Year

plt.figure(figsize=(20, 10))
sns.barplot(data=car_year_counts, x='Car_Name', y='Year', hue='Count', dodge=False)
plt.title('Number of Cars by Car Name and Year')
plt.xlabel('Car Name')
plt.ylabel('Year')
plt.xticks(rotation=90)
plt.legend(title='Count', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Create boxplot for Selling_Price column

plt.figure(figsize=(12, 8))
sns.boxplot(data=data['Selling_Price'])
plt.title('Boxplot of Selling_Price Column')
plt.show()
```

```python
# Count the occurrences of each fuel type
selling_type_counts = data['Selling_type'].value_counts()

# Plot a pie chart
plt.figure(figsize=(10, 10))
plt.pie(selling_type_counts, labels=selling_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of selling types')
plt.axis('equal')  
plt.show()
```

```python
# Count the occurrences of each fuel type
Transmission_counts = data['Transmission'].value_counts()

# Plot a pie chart
plt.figure(figsize=(10, 10))
plt.pie(Transmission_counts, labels=Transmission_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Transmissions')
plt.axis('equal')  
plt.show()
```

```python
# distplot of the price column

sns.set_style("whitegrid")
plt.figure(figsize=(7, 5))
sns.distplot(data.Present_Price)
plt.show()
```

```python
#HISTOGRAM

plt.figure(figsize=(10, 6))
sns.histplot(data['Present_Price'], bins=20, kde=True)
plt.title('Distribution of Present Price')
plt.xlabel('Present Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data['Driven_kms'])
plt.title('Violin Plot of Driven kms')
plt.xlabel('Present Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Year'], bins=20, kde=True)
plt.title('Distribution of years')
plt.xlabel('year')
plt.ylabel('Frequency')
plt.show()
```

```python
#Scatterplot

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Present_Price', y='Selling_Price')
plt.title('Scatter Plot of Present Price vs. Selling Price')
plt.xlabel('Present Price')
plt.ylabel('Selling Price')
plt.show()
```
